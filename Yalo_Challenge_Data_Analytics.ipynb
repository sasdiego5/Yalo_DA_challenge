1. Calculate the total products and revenue sold over time by quarter and identify the month where the revenue sold was 10% above the average.

```python
%%bigquery df_q1_a --project first-discovery-220520

WITH
quarter_sales AS (
SELECT DATE_TRUNC(date , QUARTER) as quarter,
       SUM(sale_dollars) as total_revenue,
       SUM(bottles_sold) as total_products
FROM `bigquery-public-data.iowa_liquor_sales.sales`
WHERE sale_dollars > 0 and bottles_sold > 0
GROUP BY quarter
)

SELECT * FROM quarter_sales ORDER BY total_sales DESC, total_bottles DESC;
```

```python
# total products and revenue over time
df_q1_a
```

```python
%%bigquery df_q1_b --project first-discovery-220520

WITH

month_sales AS (
  SELECT DATE_TRUNC(date , MONTH) as month,
         SUM(sale_dollars) as total_revenue
  FROM `bigquery-public-data.iowa_liquor_sales.sales`
  WHERE sale_dollars > 0 and bottles_sold > 0
  GROUP BY month
) ,

avg_month_sales AS (
  SELECT AVG(total_sales) as avg_revenue,
  AVG(total_sales) * 1.10 as avg_above_10_pct
  FROM month_sales
) ,

month_above_avg AS(
  SELECT month,
         total_revenue,
  FROM month_sales
  WHERE total_revenue > (SELECT avg_above_10_pct FROM avg_month_sales)
)

SELECT * FROM month_above_avg ORDER BY total_revenue DESC LIMIT 10;
```

2. List the counties where the amount (in dollars) of purchases transactions went over $100K.


```python
%%bigquery df_q2 --project first-discovery-220520

WITH county_sales AS(
  SELECT county,
         SUM(sale_dollars) as total_sales
  FROM `bigquery-public-data.iowa_liquor_sales.sales`
  GROUP BY county
)

SELECT county, total_sales,
FROM county_sales
WHERE total_sales > 100000
ORDER BY total_sales DESC;
```

3. Identify the top 10 stores with more revenue in sold products and the bottom stores with least revenue in sold products (apply a deduplication logic in case it’s needed).

```python
%%bigquery df_q3 --project first-discovery-220520


WITH store_sales AS (
  SELECT
    store_name,
    SUM(sale_dollars) AS total_sales
  FROM `bigquery-public-data.iowa_liquor_sales.sales`
  GROUP BY store_name
) ,

top_10_stores AS (
  SELECT *
  FROM store_sales
  ORDER BY total_sales DESC
  LIMIT 10
),

bottom_10_stores AS (
  SELECT *
  FROM store_sales
  ORDER BY total_sales ASC
  LIMIT 10
)

SELECT * FROM top_10_stores
UNION ALL
SELECT * FROM bottom_10_stores;
```
